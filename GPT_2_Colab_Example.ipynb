```python
from transformers import pipeline, set_seed
```

```python
set_seed(42)  # Asegura la reproducibilidad
generator = pipeline('text-generation', model='gpt2')
```

```python
prompts = ["Hello, I'm a language model,", "The quick brown fox jumps over", "In a distant future, humanity"]

# Generar texto para cada prompt
for prompt in prompts:
    generated_text = generator(prompt, max_length=50, num_return_sequences=1)
    print("Prompt:", prompt)
    print("Generated Text:", generated_text[0]['generated_text'])
    print("\n" + "-"*80 + "\n")
```
