```python
# Importar las bibliotecas necesarias
import numpy as np  # Importar la biblioteca NumPy para operaciones numéricas
import pandas as pd  # Importar la biblioteca pandas para manejo de datos

# Listar los archivos disponibles en el directorio "/kaggle/input"
import os  # Importar la biblioteca os para operaciones relacionadas con el sistema
for dirname, _, filenames in os.walk('/kaggle/input'):  # Recorre los archivos en el directorio especificado
    for filename in filenames:
        print(os.path.join(dirname, filename))  # Imprimir la ruta completa de cada archivo

# Cargar los datos de entrenamiento desde el archivo "train.csv" en un DataFrame llamado "train_data"
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")

# Mostrar las primeras 5 filas del DataFrame "train_data" para visualizar los datos
train_data.head()

# Cargar los datos de prueba desde el archivo "test.csv" en un DataFrame llamado "test_data"
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")

# Mostrar las primeras 5 filas del DataFrame "test_data" para visualizar los datos
test_data.head()

# Importar el clasificador RandomForestClassifier de scikit-learn
from sklearn.ensemble import RandomForestClassifier  # Importar el clasificador de Bosques Aleatorios

# Seleccionar la columna "Survived" como la variable objetivo (a predecir)
y = train_data["Survived"]

# Seleccionar las características (columnas) que se utilizarán para predecir la supervivencia
features = ["Pclass", "Sex", "SibSp", "Parch"]

# Aplicar "one-hot encoding" a las columnas categóricas como "Sex"
X = pd.get_dummies(train_data[features])  # Transformar las características categóricas en variables binarias
X_test = pd.get_dummies(test_data[features])

# Crear un modelo RandomForestClassifier con 100 árboles y profundidad máxima de 5
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)

# Ajustar el modelo a los datos de entrenamiento
model.fit(X, y)

# Realizar predicciones en los datos de prueba
predictions = model.predict(X_test)

# Crear un DataFrame con las predicciones y el identificador de pasajero
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})

# Guardar las predicciones en un archivo CSV llamado "submission.csv"
output.to_csv('submission.csv', index=False)

# Imprimir un mensaje de confirmación
print("Your submission was successfully saved!")
```
