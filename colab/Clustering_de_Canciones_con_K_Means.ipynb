```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Datos ficticios de canciones
datos_canciones = {
    'Duracion': [210, 180, 240, 300, 270, 220, 200, 230, 250, 190],
    'Energia': [0.8, 0.7, 0.9, 0.65, 0.78, 0.85, 0.75, 0.82, 0.77, 0.7],
    'Bailabilidad': [0.7, 0.8, 0.65, 0.6, 0.7, 0.75, 0.8, 0.85, 0.75, 0.7]
}

# Convertir a un array de NumPy para el algoritmo K-Means
X = np.array(list(zip(datos_canciones['Duracion'], datos_canciones['Energia'], datos_canciones['Bailabilidad'])))
```

```python
# Aplicar K-Means
kmeans = KMeans(n_clusters=3)  # Elegimos 3 grupos para este ejemplo
kmeans.fit(X)

# Obtener las etiquetas de los grupos
etiquetas = kmeans.labels_
```

```python
# AÃ±adir las etiquetas al DataFrame
datos_canciones['Grupo'] = etiquetas

# Crear un DataFrame con los datos
df_canciones = pd.DataFrame(datos_canciones)

# Graficar los resultados
sns.pairplot(df_canciones, hue='Grupo', palette='bright')
```
