```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
diametros = [[6], [8], [10], [14], [18]]
precios = [[7], [9], [13], [17.5], [18]]
```

```python
# Creamos el modelo de regresión lineal
modelo = LinearRegression()

# Entrenamos el modelo con nuestros datos de diámetros y precios
modelo.fit(diametros, precios)
```

```python
# Hacer predicciones para nuestros diámetros
precios_predichos = modelo.predict(diametros)

# Dibujar los puntos de datos y la línea de regresión
plt.figure()
plt.title('Precio y diámetro de pizzas')
plt.xlabel('Diámetro (cm)')
plt.ylabel('Precio (euros)')
plt.plot(diametros, precios, 'k.')  # 'k.' dibuja puntos negros para nuestros datos reales
plt.plot(diametros, precios_predichos, 'b-')  # 'b-' dibuja una línea azul para nuestra predicción
plt.axis([0, 25, 0, 25])
plt.grid(True)
plt.show()
```
