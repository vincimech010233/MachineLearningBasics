```python
# Importa las bibliotecas necesarias
import gym
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML, display, clear_output

# Define la clase del agente
class RandomAgent:
    def __init__(self, env):
        self.env = env

    def get_action(self):
        return self.env.action_space.sample()

# Crea el entorno y el agente
env = gym.make('CartPole-v1')
agent = RandomAgent(env)

# Función para correr y visualizar un episodio
def run_and_display_episode(env, agent):
    state = env.reset()
    done = False
    frames = []  # Lista para almacenar los frames
    while not done:
        frames.append(env.render(mode='rgb_array'))  # Guarda el frame actual
        action = agent.get_action()
        state, reward, done, info = env.step(action)
    env.close()
    return frames

# Corre el episodio y obtiene los frames
frames = run_and_display_episode(env, agent)

# Crea una animación con los frames y la muestra
fig, ax = plt.subplots()
patch = plt.imshow(frames[0])
plt.axis('off')
def animate(i):
    patch.set_data(frames[i])
anim = animation.FuncAnimation(plt.gcf(), animate, frames=len(frames), interval=50)
video = anim.to_jshtml()
display(HTML(video))
```
